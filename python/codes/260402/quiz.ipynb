```python
# Q1. 윤년
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
year = int(input())
if ((not (year%4)) and (year%100)) or not (year % 400):
    print(1)
else: print(0)
```

```python
# Q2. 사분면 고르기
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
x, y = int(input()),int(input())
if x>0 and y>0:
    print(1)
if x<0 and y>0:
    print(2)
if x<0 and y<0:
    print(3)
if x>0 and y<0:
    print(4)
```

```python
# Q3. 알람 시계
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
H, M = map(int, input().split())
seted_time = H * 60 + M - 45
if seted_time < 0:
    seted_time+=(24*60)
print(f"{seted_time // 60} {seted_time % 60}") 
```

```python
# Q4. 오븐 시계
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
A, B = map(int, input().split())
C = int(input())
seted_time = (A * 60 + B + C) % (24*60)
print(f"{seted_time // 60} {seted_time % 60}") 
```

```python
# Q5. 주사위 세개
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
dices = list(map(int, input().split()))
result = [0] * 6
for i in dices:
    result[i-1]+=1
for i in range(5, -1, -1):
    if result[i] == max(result):
        if result[i] == 3:
            print((1+i)*1000+10000)
            break
        if result[i] == 2:
            print((1+i)*100+1000)
            break
        if result[i] == 1:
            print((1+i)*100)
            break
        
```

```python
# Q7. 구구단
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
N = int(input())
for i in range(1,10):
    print(f"{N} * {i} = {N*i}")
```

```python
# Q8. 합
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
sum_num = 0
n = int(input())
print(sum([(sum_num + i) for i in range(1,n+1)]))
```

```python
# Q9. 영수증
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
X, N, normal_pay= int(input()), int(input()), 0
for _ in range(N):
    a,b = map(int, input().split())
    normal_pay+= (a*b)
if X == normal_pay:
    print("Yes")
else:
    print("No")

```
