```python
i, sum_result=0, 0
while(True):
    sum_result+=i
    i+=1
    print(sum_result)
    if sum_result >= 1000:
        break
```

```python
i =0 # 1에서 100까지 홀수만
while(True):
    i+=1
    if i % 2 == 0:
        continue
    if i >= 100:
        break
    print(i)
```

```python
#1에서 100까지 3의 배수와 4의 배수 출력
i=1
while(True):
    i+=1
    if (i % 4 == 0) | (i % 3 == 0):
        print(i)
        continue
    if i >= 100:
        break
    
```

```python
num = [5,15,6,7,25]
#1) 10보다 작은 수는 건너띄고 출력
#2) 20보다 큰 수 나오면 멈추고 출력
for i in num:
    if i <10: continue
    if i >20:
        print(i)
        break
```
