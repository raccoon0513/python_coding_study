```python
#cospro 2급 1번

def Cnt_size(shirt_size):
    shirt_size_list = ["XS", "S", "M", "L", "XL", "XXL"]
    result = [0] * 6
    for i in shirt_size:
        result[shirt_size_list.index(i)]+=1
    return result

shirt_size = ["XS", "S", "L", "L", "XL", "S"]

print(Cnt_size(shirt_size))
```

```python
#cospro 2급 2번
def solution(price, grade):
    grades = {
        "S":5,
        "G":10,
        "V":15
        }
    return int(price * discount_rate(grades[grade]))
def discount_rate(rate):
    return 1 - (rate/100)

print(solution(2500, "V"))
print(solution(96900, "S"))
```

```python
#cospro 2급 3번
month_list = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
def solution(start_month, start_day, end_month, end_day):
    return total_day(end_month, end_day) - total_day(start_month, start_day)
def total_day(month, day):
    return sum([month_list[i] for i in range(month)])+day
print(solution(1,2,2,2)) # 31
```

```python
#cospro 2급 4번
# 최다등장 횟수가 최소 등장 횟수의 몇 배인가
# arr = [1,2,3,3,1,3,3,2,3,2]
arr  = [1, 1, 2, 2, 2, 3, 3, 3, 3, 3]
counter = {1:1}
for i in arr:
    #삼항연산자에는 대입연산이 불가능합니다.
    if (i in counter) : 
        counter[i]+=1
    else:
        counter[i]=1
print(max(counter.values()), min(counter.values()))
```

```python
# 6번 369 박수치기
N = int(input())
for i in range(1, N+1):
    isin = False
    # any() 함수 : 리스트 안 조건중 하나라도 참일시 참
    # [매개변수 for 매개변수 in [3,6,9]] 형식에 따라 리스트컴프리헨션
    # str(x) in str(i) 조건에 따라 True 혹은 False로 저장되는 리스트가 생성됨
    if any(str(x) in str(i) for x in [3,6,9]):
        isin = True
    print(f"{i}{', 짝' if isin else''}")
```
