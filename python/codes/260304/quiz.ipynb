```python
# 문제 1 등변사다리의 높이
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
import math
user_input = input()
r, d1, d2 = user_input.split()
r = int(r)
d1 = int(d1)
d2 = int(d2)
h=0
h = ( r**2 - (abs(d1-d2) / 2 )**2 ) *4
print(int(h))
```

```python
# 문제 2 택시 요금
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
user_input = input()
A,B,C,D = map(int, user_input.split())
pay = 0
if D < C :
	pay = D * A
else : 
	pay = (C * A) + ((D-C) * B)
print(int(pay))
```

```python
# 문제 3 바느질
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
import math
count = int(input())
arr = []
for i in range(count):
	arr.append(map(list(int, input().split())))
	
def remove_test(arr, remove_index):
	test_arr = arr
	test_arr = test_arr.remove(remove_index)
	max = 0
	total_distance =0
	for i in range(len(test_arr)-1):
		distance = abs(test_arr[i][0] - test_arr[i+1][0]) + abs(test_arr[i][1] - test_arr[i+1][1])
		total_distance += distance
		if distance > max:
			max = distance
	return max, total_distance

result_arr = []
for i in range(count):
	result_arr.append(remove_test(arr, i)[0])
print(remove_test(result_arr.index(result_arr.max()))[1])
	

	
```
