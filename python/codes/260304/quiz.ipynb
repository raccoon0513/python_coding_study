```python
# 문제 3 바느질
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
import math
count = int(input())
arr = []
for i in range(count):
	arr.append(list(map(int, input().split())))

def manhattan_dist(p1, p2):
	x1, y1 = p1
	x2, y2 = p2
	return abs(x1 - x2) + abs(y1 - y2) 
	
def remove_test(arr):
	total_distance =0
	for i in range(len(arr)-1):
		total_distance += arr
	return total_distance

full_distance = 0
for i in range(len(arr)-1):
	full_distance += manhattan_dist(arr[i], arr[i+1])

results = []
for i in range(1, count-1):
	before = manhattan_dist(arr[i], arr[i-1]) + manhattan_dist(arr[i], arr[i+1])
	after = manhattan_dist(arr[i-1], arr[i+1])
	results.append(after - before)

print(full_distance + min(results))
```

```python
# 문제 5 조별 과제
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
N, G, B = map(int, input().split())

g_progression, b_progression = 0, 0
def exchange():
    global g_progression, b_progression
    g_progression, b_progression = b_progression, g_progression

for i in range(N):
    
    g_progression+=G
    b_progression+=B
    if g_progression < b_progression:
        exchange()
    elif g_progression == b_progression:
        g_progression +=1

print(g_progression, b_progression)
```

```python
# 문제 7 최소 연산으로 배열 정렬하기
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
arr_len = int(input())
arr = list(map(int, input().split()))
arr_sorted = sorted(arr)

false_index_arr = []
for i in range(len(arr)):
    if arr[i] != arr_sorted[i]:
        false_index_arr.append(i)

false_len = len(false_index_arr)


def checker():
    #전부맞으면 통과
    if false_len == 0:
        print("yes")
        return
    
    #안맞으면 리턴
    elif(false_len==6):
        print("no")
        return

    
    
    else:
        start = false_index_arr[0]
        end = false_index_arr[-1]

         # 두개면 스왑
        if false_len == 2:
            print("yes")
            print(f"swap {start+1} {end+1}")
            return
        #리버스체크
        reverse_able = True
        for i in range(start, end):
            if arr[start] != arr_sorted[end]: 
                reverse_able = False
                break
        if reverse_able:
            print("yes")
            print(f"reverse {start+1} {end+1}")
            return
        
       
        #아니면 no
        print("no")
        return
        

checker()
```
