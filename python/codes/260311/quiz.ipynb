```python
# Q2. 의좋은 형제
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean

class bro:
    def __init__(self, food):
        self.food = food

    def give_cal(self, count, other):
        self.food = self.food - count
        other.take(count)

    def give(self, other):
        if(self.food % 2):
            self.give_cal(self.food//2+1, other)
        else:
            self.give_cal(int(self.food/2), other)

    def take(self, count):
        self.food = self.food + count

N1, N2 = map(int ,input().split())
D = int(input())
bros = []
bros.append(bro(N1)) # 얘가 먼저 줌
bros.append(bro(N2))

for i in range(D):
    bros[i%2].give(bros[(i+1)%2])

print(bros[0].food, bros[1].food)
#출력은 bro0, bro1 순
```

```python
# Q7. 합 계산기
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
def cul(a, math_sign, b):
    a,b =int(a),int(b)
    if math_sign == "+":
        return a + b
    if math_sign == "-":
        return a - b
    if math_sign == "/":
        return a // b
    if math_sign == "*":
        return a * b

T = int(input())
result = 0
for i in range(T):
    a, math_sign, b = input().split()
    result+= cul(a, math_sign, b)
print(result)
```
