```python
# Q2. [KOI 2017] 딱지놀이
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
def count_values(arr : list):
    result = 0
    for i in range(1,4+1):
        result += ((1000 ** (i-1))*arr.count(i))
    return result

def compare(a, b):
    a_result, b_result = count_values(a), count_values(b)
    if a_result == b_result:
        print("D")
    else :
        print("A" if a_result>b_result else "B")

N = int(input())
for i in range(N):
    a = list(map(int, input().split()))[1:]
    b = list(map(int, input().split()))[1:]
    compare(a,b)
```

```python
import numpy as np

T = int(input()) #테스트 케이스 개수
for _ in range(T):
    N, K = map(int,input().split())#공터크기, 땅의크기
    arr = np.array(
        [list(map(int, input().split())) for _ in range(N)]
        )

    min_num = float('inf')
    for i in range(N+1-K):
        for j in range((N+1-K)):
            temp = np.sum(arr[i:i+K, j:j+K])
            if temp < min_num:
                min_num = temp
    print(min_num)
```

```python
# Q7. 완벽한 햄버거 만들기
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
N = int(input())
not_sort_check = False
arr = list(map(int, input().split()))
taste = 0
max_value_index = arr.index(max(arr))

if max_value_index != 0:
    for i in range(max_value_index):
        if arr[i]>arr[i+1]:
            not_sort_check = True
            break
        taste += arr[i]
if max_value_index != N-1:
    for i in range(max_value_index, N-1):
        if arr[i]<arr[i+1]:
            not_sort_check = True
            break
        taste += arr[i]
taste -=arr[max_value_index]
print(0 if not_sort_check else sum(arr))
```

```python
# Q8. 고장난 컴퓨터
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
N, c = map(int, input().split())
t = list(map(int, input().split()))

count = 1
if N == 1:
    print(count)
else:
    for i in range(1, N)[::-1]:
        if t[i] - t[i-1] > c:
            break
        count += 1
    print(count)
```

```python
# Q9. 뱀이 지나간 자리
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
N, M = map(int, input().split()) #세로, 가로

isInRight = False
isCrossing = False
while N!=0:
    if isCrossing:
        print(*["#" for _ in range(M)])
        isInRight = not isInRight
    else :     
        print(*["#" if not isInRight else ""]+["#" if isCrossing else "." for _ in range(M)]+["#" if isInRight else ""], sep="")
        

    N-=1
```

```python
# Q9. 뱀이 지나간 자리
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
N, M = map(int, input().split()) #세로, 가로
for i in range(N):
    if i % 2 == 0:
        print("#"*M)
    elif i % 4 == 1:
        print(("."*(M-1)) + "#")
    elif i % 4 == 3 :
        print("#"+ ("."*(M-1)) )
```

```python
# Q10. 거스름 돈
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
N =int(input())
def coin_count(money, money_arr, count=0, index=0):
    if index == len(money_arr):
        return count
    count += (money // money_arr[index])
    money %= money_arr[index]
    return coin_count(money, money_arr, count, index+1)
money_arr = [40, 20, 10, 5, 1]
print(coin_count(N, money_arr))
```
