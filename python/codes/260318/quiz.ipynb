```python
# Q1. 원 접기
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
import math
AM, CM, BM = map(int, input().split())
a, b = AM*BM , CM
gcd = math.gcd(a, b)
a = AM*BM // gcd
b =  CM // gcd
print(a, b)
```

```python
# Q2. 구구단을 외자 구구단을 외자
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
dan = int(input("단 입력"))
print(f"====={dan}단====")
for i in range(1,10):
    print(f"{dan} x {i} = {dan*i}")
```

```python
absent = [2,5]
no_book = [7]
for i in range(1, 10+1):
    
    if (i in absent):
        continue
    print(f"출석번호 : {i}")
    if(i in no_book):
        print("교무실로 따라오세요~~")
        break
    print("네\n")
```
