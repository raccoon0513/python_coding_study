```python
class account:
    def __init__(this, money=0):
        this.money = money
        this.open_account()
        this.menu()
    #계좌 개설
    def open_account(this):
        print("계좌 개설됨")

    # 잔액 출력
    def print_balance(this):
        print(f"잔액은 {this.money}원 입니다.")
    # 입금
    def depost(this):
        try:
            money = int(input("입금할 금액을 입력해주세요"))
            this.money += money
            this.print_balance()
        except ValueError as e:
            print(e)
            print("잘못된 값입니다")

    # 출금
    def withdraw(this):
        try:
            this.print_balance()
            money = int(input("출금할 금액은"))
            if this.money < money:
                print("한도초과입니다")
            else :
                this.money -= money
                print("출금되었습니다.")
                this.print_balance()
                return money
        except ValueError as e:
            print(e)
            print("잘못된 값입니다")
        
    def menu(this):
        while(True):
            try:
                index = int(input("메뉴를 선택하세요\n" \
                "1. 입금\n" \
                "2. 출금\n" \
                "3. 잔액 확인\n" \
                "4. 종료"))-1
                defs = [
                    this.depost,
                    this.withdraw,
                    this.print_balance
                ]
                if(index == 3):
                    print("안녕히가세용")
                    break
                defs[index]()
            except Exception as e:
                print(f"{index}는 잘못된 값입니다 : {e}")
            
            
ac = account(500)
```
