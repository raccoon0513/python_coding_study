```python
# Q1. 막대기
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
N = int(input())
count,h_max, h = 0, 0, []
for i in range(N):
    h.append(int(input()))
h_max_index = h.index(max(h))
if h_max_index == max(h)-1:
    print(1)
else:
    for i in range(len(h),h_max_index):
        if i > h_max:
            count+=1
            h_max = i
    print(count)
    print(f"{i}")
```
