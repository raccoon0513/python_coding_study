```python
# Q1. 막대기
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
N = int(input())
count,h_max, h = 0, 0, []
for i in range(N):
    h.append(int(input()))
h_max_index = h.index(max(h))

for i in range(h_max_index,len(h))[::-1]:
    if h[i] > h_max:
        count+=1
        h_max = h[i]
print(count)
```

```python
# Q2. 소금물의 농도 구하기
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
# 기본 7%, N = 7% 소금물 Ng, M = 물 Mg
N, M = map(int, input().split())
salt = N * 0.07
print(f"{int(((salt/(M+N))*100)*100)/100:.2f}")
```

```python
# Q10. 시험 성적
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
score = int(input())
grades = ["A","B","C","D","F"]
cut = [90, 80, 70, 60, 0]

for i in range(len(grades)):
    if score >= cut[i]:
        print(grades[i])
        break
```
